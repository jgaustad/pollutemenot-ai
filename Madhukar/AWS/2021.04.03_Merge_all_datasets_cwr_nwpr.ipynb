```python
# Baseline
df_baseline = pd.read_pickle("cwr_nwpr_dataset")
df_baseline.drop_duplicates(subset='da_number', keep="first", inplace=True)
print(df_baseline.shape[0])

# NHD
df_nhd_200m = pd.read_pickle("cwr_nwpr_200m_nhd_variables_extracted_stripped")
df_nhd_1000m = pd.read_pickle("cwr_nwpr_1000m_nhd_variables_extracted_stripped")
df_nhd_2500m = pd.read_pickle("cwr_nwpr_2500m_nhd_variables_extracted_stripped")

df_nhd_200m.drop_duplicates(subset='da_number', keep="first", inplace=True)
df_nhd_1000m.drop_duplicates(subset='da_number', keep="first", inplace=True)
df_nhd_2500m.drop_duplicates(subset='da_number', keep="first", inplace=True)

print(df_nhd_200m.shape[0])
print(df_nhd_1000m.shape[0])
print(df_nhd_2500m.shape[0])

# SSURGO
df_ssurgo = pd.read_pickle("2021.04.02_ssurgo_cwr_nwpr_variables_full")
df_ssurgo.drop_duplicates(subset='da_number', keep="first", inplace=True)

print(df_ssurgo.shape[0])

# NWI
df_nwi_200m = pd.read_pickle("nwi_extracted_features_cwr_nwpr_200X200m")
df_nwi_1000m = pd.read_pickle("nwi_extracted_features_cwr_nwpr_1000X1000m")
df_nwi_2500m = pd.read_pickle("nwi_extracted_features_cwr_nwpr_2500X2500m")

df_nwi_200m.drop_duplicates(subset='da_number', keep="first", inplace=True)
df_nwi_1000m.drop_duplicates(subset='da_number', keep="first", inplace=True)
df_nwi_2500m.drop_duplicates(subset='da_number', keep="first", inplace=True)

print(df_nwi_200m.shape[0])
print(df_nwi_1000m.shape[0])
print(df_nwi_2500m.shape[0])

# srtm, slope, jrc
df_srtm_200m = pd.read_pickle("df_cwr_nwpr_srtm_slope_variables_200m")
df_srtm_1000m = pd.read_pickle("df_cwr_nwpr_srtm_slope_variables_1000m")
df_srtm_2500m = pd.read_pickle("df_cwr_nwpr_srtm_slope_variables_2500m")

df_srtm_200m.drop_duplicates(subset='da_number', keep="first", inplace=True)
df_srtm_1000m.drop_duplicates(subset='da_number', keep="first", inplace=True)
df_srtm_2500m.drop_duplicates(subset='da_number', keep="first", inplace=True)

print(df_srtm_200m.shape[0])
print(df_srtm_1000m.shape[0])
print(df_srtm_2500m.shape[0])

# county
df_county = pd.read_pickle("2021.04.01_cwr_nwpr_us_county_raw")
df_county.drop_duplicates(subset='da_number', keep="first", inplace=True)
print(df_county.shape[0])

# HUC
df_huc4 = pd.read_pickle("2021.04.01_cwr_nwpr_huc4")
df_huc6 = pd.read_pickle("2021.04.01_cwr_nwpr_huc6")

df_huc4.drop_duplicates(subset='da_number', keep="first", inplace=True)
df_huc6.drop_duplicates(subset='da_number', keep="first", inplace=True)

print(df_huc4.shape[0])
print(df_huc6.shape[0])
```

```python
df1 = df_baseline
df2 = df_ssurgo

df3 = df_srtm_200m
df4 = df_nhd_200m
df5 = df_nwi_200m

df6 = df_srtm_2500m
df7 = df_nhd_2500m
df8 = df_nwi_2500m

df9 = df_county

# df10 = df_nearest_wotus

df11 = df_huc4
df12 = df_huc6

df13 = df_srtm_1000m
df14 = df_nhd_1000m
df15 = df_nwi_1000m

df_merged = pd.merge(df1, df2, on="da_number")
df_merged = pd.merge(df_merged, df3, on="da_number")
df_merged = pd.merge(df_merged, df4, on="da_number")
df_merged = pd.merge(df_merged, df5, on="da_number")
df_merged = pd.merge(df_merged, df6, on="da_number")
df_merged = pd.merge(df_merged, df7, on="da_number")
df_merged = pd.merge(df_merged, df8, on="da_number")
df_merged = pd.merge(df_merged, df9, on="da_number")
df_merged = pd.merge(df_merged, df11, on="da_number")
df_merged = pd.merge(df_merged, df12, on="da_number")
df_merged = pd.merge(df_merged, df13, on="da_number")
df_merged = pd.merge(df_merged, df14, on="da_number")
df_merged = pd.merge(df_merged, df15, on="da_number")
set(df_merged.columns)

```
