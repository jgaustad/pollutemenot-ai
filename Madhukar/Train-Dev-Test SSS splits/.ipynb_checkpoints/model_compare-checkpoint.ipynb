```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('combined_regular_clean_with_ssurgo_variables.csv')
df.head()
```

```python
df_gbm = pd.concat([pd.read_csv('lightgbm_dev_predictions.csv'), pd.read_csv('lightgbm_test_predictions.csv')])
df_cnn = pd.concat([pd.read_csv('dev_predictions_cnn_1.csv'), pd.read_csv('test_predictions_cnn_1.csv')])

display(df_gbm)
display(df_cnn)
```

```python
df_cnn = df_cnn[['da_number', 'cwa_determination', 'predicted', 'predicted_unround']]
df_cnn.columns = ['da_num', 'cwa_det', 'cnn_pred', 'cnn_pred_unround']
df_cnn
```

```python
df_gbm = df_gbm[['da_number', 'cwa_determination', 'prediction', 'predict_proba']]
df_gbm.columns = ['da_num', 'cwa_det', 'gbm_pred', 'gbm_pred_proba']
df_gbm
```

```python
df_preds = pd.merge(df_cnn, df_gbm, on = 'da_num', how = 'outer')
print(df_preds.shape)
df_preds.head()
```

```python
df_full = pd.merge(df_preds, df, left_on = 'da_num', right_on ='da_number', how='left')
print(df_full.shape)
```

```python
df_full['correct_cnn'] = df_full.cnn_pred == df_full.cwa_det_x
df_full['correct_gbm'] = df_full.gbm_pred == df_full.cwa_det_x
df_full['unique_cnn_correct'] = df_full.apply(lambda x: True if x['correct_cnn'] == True and x['correct_gbm'] == False else False, axis = 1)
df_full['unique_gbm_correct'] = df_full.apply(lambda x: True if x['correct_gbm'] == True and x['correct_cnn'] == False else False, axis = 1)

df_full.head()
```
